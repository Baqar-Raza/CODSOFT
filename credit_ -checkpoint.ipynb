```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
X=train[["amt","zip","lat","long","job","unix_time"]]
```

```python
vif_data = pd.DataFrame() #converting the data into df
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
  
print(vif_data)
```

```python
train=train.drop("unix_time", axis=1)
train
```

```python
Y=train.loc[:,"is_fraud"]
X=train.drop("is_fraud",axis=1)
Y
X
```

```python
test.job=pd.Categorical(test.job).codes
X_test=test[["amt","zip","lat","long","job"]]
X_test
```

```python
Y_test=test.loc[:,"is_fraud"]
Y_test
```

```python
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression  
classifier= LogisticRegression(random_state=0)  
classifier.fit(X,Y) 
```

```python
Y_Pred=classifier.predict(X_test)
print(classification_report(Y_test,Y_Pred))
```

```python
from sklearn.metrics import accuracy_score

# Calculate the accuracy
accuracy = accuracy_score(Y_test, Y_Pred)
print(f"Accuracy: {accuracy:.2f}")
```
